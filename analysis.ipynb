```python
df = pd.read_csv('/kaggle/input/goldman-sachs-all-time-stock-data/goldmansachs.csv')
```

```python
def s_search(user):
    years = df['date']
    
    year = df['date'].str[:4]
    
    n_notebook = df[user == year]
    
    return n_notebook
```

```python
def user_search():
  
    while True:
        user_option = input("Enter year, 1999 - 2026\n")
        if (int(user_option) >= 1999 and int(user_option) <= 2026):
            break
        print('Please enter a year, 1999 - 2026')
        
    
    return s_search(user_option)
```

```python

def a_graph() -> pd.DataFrame:
    lst = []
    info_lst = []
    years = df['date'].str[:4]
    for year in years:
        if not year in lst:
            lst.append(year)
            info_lst.append([year, sum(s_search(year)['open'])/ len(s_search(year)), min(s_search(year)['low']), sum(s_search(year)['close'])/ len(s_search(year)), max(s_search(year)['high']), sum(s_search(year)['volume']) / len(s_search(year))])
    
    n_df = pd.DataFrame(info_lst, columns = ['Year', 'Avg_Open', 'Abs_Low', 'Avg_Close', 'Abs_High', 'Avg_Volume'])
    
    return n_df.to_csv("Analysis_file.csv")
a_graph()
```
